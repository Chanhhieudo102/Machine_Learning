## **2. Read data**

```python
data = pd.read_csv('/content/Iris.csv')
data.head()
```

### **Prepare data**

```python
x_data = data[['PetalLengthCm', 'PetalWidthCm']].to_numpy()

for i in range(len(x_data)):
    print(f"Point {i}: {x_data[i]}")
```

```python
y_data = data['Species'].to_numpy()

for i in range(len(y_data)):
    print(f"Label of point {i}: {y_data[i]}")
```

```python
x_test = np.array([2.4, 0.8])
print(f"Test point: {x_test}")
print(x_test.shape)
x_test = x_test.reshape(1, 2)

print(f"Test point: {x_test}")
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_data_scaled = scaler.fit_transform(x_data)
x_test_scaled = scaler.transform(x_test)
### WRITE YOUR CODE HERE


print(f"Data before scaling: \n{x_data}")
print(f"Data after scaling: \n{x_data_scaled}")
print(f"Test data before scaling: \n{x_test}")
print(f"Test data after scaling: \n{x_test_scaled}")
```

##  **3. Implement KNN**

#### Compute distances

```python
distances = (x_data_scaled - x_test_scaled)
distances = np.sum(distances**2)
distances
```

```python
distances = (x_data_scaled - x_test_scaled) ** 2
distances = np.sum(distances, axis=1)
distances = np.sqrt(distances)

### WRITE YOUR CODE HERE

for i in range(len(distances)):
    print(f"Distance from point {i} to test point: {distances[i]:.4f}")
```

#### Sort the distances

```python

sort_distances = np.argsort(distances)
### WRITE YOUR CODE HERE

print(f"Indices of points sorted by distance: {sort_distances}")
```

#### Get the top k nearest neighbors

```python
k = 3
top_k = sort_distances[:k]

### WRITE YOUR CODE HERE

print(f"Top {k} nearest neighbors: {top_k}")
```

```python
print(distances[top_k])
```

```python
print(y_data[top_k])
```

#### Vote for the class

```python
from collections import Counter
counter = Counter(y_data[top_k])
predicted_label = counter.most_common(1)[0][0]


### WRITE YOUR CODE HERE

print(f"The predicted label is {predicted_label}")
```

## **4. Visualization**

```python
nearest_neighbors = x_data[top_k]
nearest_neighbors
plt.figure(figsize=(8, 6))
for label in np.unique(y_data):
  mask = (y_data == label)
  plt.scatter(x_data[mask, 0], x_data[mask, 1], label = f'label {(label)}')

plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title(f'KNN Prediction (k={k})')
plt.legend()
plt.grid(True)
plt.show()
```
